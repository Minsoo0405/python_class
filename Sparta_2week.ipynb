### 1) 수익률 구해보기

```python
# 쉽게 해보기 : 우선 1,000씩 다 나눠보기
df = pd.DataFrame()

us500 = fdr.DataReader('US500', '2017')
df['Price'] = us500['Close']/1000 -1

df
```

```python
# 1,000이 아니라 처음 값으로 나눠보기

df = pd.DataFrame()

us500 = fdr.DataReader('US500', '2017')
df['Price'] = us500['Close']/us500['Close'].iloc[0] -1

df
```

```python
# 함수로 만들어보기

def make_rate(code_name):
    us500 = fdr.DataReader(code_name, '2017')
    return us500['Close']/us500['Close'].iloc[0] -1 
```

```python
df = pd.DataFrame()

df['META'] = make_rate('META')
df['GOOG'] = make_rate('GOOG')
df['US500'] =make_rate('US500')
df.plot()
```

### 1) 최종 수익률 구하는 함수 정의

```python
def final_rate(code_name):
    return make_rate(code_name).iloc[-1]
```

# 8. S&P 500 기업 전체 (2)

```python
sp500 = fdr.StockListing('SP500')
df = sp500[['Symbol', 'Name']].head(10)
df['rate'] = df['Symbol'].apply(final_rate)
df = df.sort_values(by='rate', ascending=False)
df
```

### 1) final_rate 함수 재정의

```python
def final_rate(code_name):
    try:
        return make_rate(code_name).iloc[-1]
    except:
        return (-1000)
```

### 2) 100개 기업 수익률 보기

```python
sp500 = fdr.StockListing('SP500')
df = sp500[['Symbol', 'Name']].head(100)
df['rate'] = df['Symbol'].apply(final_rate)
df = df.sort_values(by='rate', ascending=False)

df.set_index('Name', inplace=True)
df
```

# 9. 2주차 끝 & 숙제

### 1) S&P 500 기업 100개 중 수익률 best 10 기업을 찾고, 바 그래프를 그려보세요!

```python
# 수익률 best 10 기업 찾기
sp500 = fdr.StockListing('SP500')
df = sp500[['Symbol', 'Name']].head(100)
df['rate'] = df['Symbol'].apply(final_rate)
df = df.sort_values(by='rate', ascending=False).head(10)

df.set_index('Name', inplace=True)
df
```

```python
# best 10 기업 바 그래프 그리기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x=df.index, y=df['rate'])
plt.xticks(rotation=90)
plt.title('Best 10 out of 100 S&P 500 companies')
plt.show()
```
